selection sort (선택 정렬, 교환법)  
: 선택 정렬은 i 번째 데이터를 선택해서 나머지 데이터(j) 와 비교하여 정렬한다.
  정렬할 데이터가 n개일 경우 회전수는 n-1 번이 된다. => 5개 데이터 : 회전 4번

```python
def selectionAsc(data) : # 오름차순
    for i in range(len(data)):
        for j in range(i+1 , len(data)):
            if data[i] > data[j] :
                data[i] , data[j] = data[j], data[i]
    return data
```

```python
def selectionDsc(data) : # 내림차순
    for i in range(len(data)): 
        for j in range(i+1 , len(data)):
            if data[i] < data[j] :
                data[i] , data[j] = data[j], data[i]
    return data
```

```python
data = []
n = 0
while n != 999 :
    n = int(input('숫자를 입력하세요: '))
    data.append(n)
data.pop()
print('오름차순\n{}'.format(selectionAsc(data)))
print('내름차순\n{}'.format(selectionDsc(data)))

    
```
