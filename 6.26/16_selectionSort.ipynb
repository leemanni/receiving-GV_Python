selection sort ( , )  
:   i     (j)   .
    n   n-1  . => 5  :  4

```python
def selectionAsc(data) : # 
    for i in range(len(data)):
        for j in range(i+1 , len(data)):
            if data[i] > data[j] :
                data[i] , data[j] = data[j], data[i]
    return data
```

```python
def selectionDsc(data) : # 
    for i in range(len(data)): 
        for j in range(i+1 , len(data)):
            if data[i] < data[j] :
                data[i] , data[j] = data[j], data[i]
    return data
```

```python
data = []
n = 0
while n != 999 :
    n = int(input(' : '))
    data.append(n)
data.pop()
print('\n{}'.format(selectionAsc(data)))
print('\n{}'.format(selectionDsc(data)))

    
```
